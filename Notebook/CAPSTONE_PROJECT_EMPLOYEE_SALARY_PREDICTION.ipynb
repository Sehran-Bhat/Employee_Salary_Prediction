```python
X=df[['Years_at_Company','Satisfaction_Level','Average_Monthly_Hours']]
y=df['Salary']
```

```python
!pip install scikit-learn
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
```

```python
X_train
```

```python
import joblib
joblib.dump(scaler, 'scaler.pkl')
```

```python
X_test=scaler.fit_transform(X_test)
X_test
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error,mean_absolute_error

def results(prediction):
  print("Mean absolute error on model is {}",format(mean_absolute_error(y_test,prediction)))
  print("Root mean squared error on model is {}",format(mean_squared_error(y_test,prediction)))
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
predictionlr=lr.predict(X_test)
```

```python
results(predictionlr)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
```

```python
svrmodel=SVR()
```

```python
param_gridsvr={
    'C':[0.1,1,10],
    'epsilon':[0.1,0.2,0.5],
    'kernel':['linear','poly','rbf'],#ensure valid kernel types
    'degree':[2,3,4],#only applicable if poly kernel is used

}
```

```python
gridsvr=GridSearchCV(svrmodel,param_gridsvr)
```

```python
gridsvr.fit(X_train,y_train.values.ravel())
```

```python
gridsvr.best_params_
```

```python
predictionssvr=gridsvr.predict(X_test)
```

```python
results(predictionssvr)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
rfmodel=RandomForestRegressor()
```

```python
param_gridrfr={
    'n_estimators':[2,3,4,5,6],#use a list instead of a set
    'max_depth':[5,10,15],
}
```

```python
gridrfr=GridSearchCV(estimator=rfmodel,param_grid=param_gridrfr,cv=5)
```

```python
gridrfr.fit(X_train,y_train.values.ravel())
```

```python
gridrfr.best_params_
```

```python
predictionofrfr=gridrfr.predict(X_test)
```

```python
results(predictionofrfr)
```

```python
lr
```

```python
joblib.dump(lr,'model.pkl')
```

```python
# Create a bar chart to compare the performance of the models
models = ['Linear Regression', 'SVR', 'Random Forest']
mae_scores = [mean_absolute_error(y_test, predictionlr),
              mean_absolute_error(y_test, predictionssvr),
              mean_absolute_error(y_test, predictionofrfr)]
rmse_scores = [mean_squared_error(y_test, predictionlr),
               mean_squared_error(y_test, predictionssvr),
               mean_squared_error(y_test, predictionofrfr)]

x = np.arange(len(models))
width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
rects1 = ax.bar(x - width/2, mae_scores, width, label='MAE')
rects2 = ax.bar(x + width/2, rmse_scores, width, label='RMSE')

ax.set_ylabel('Error Score')
ax.set_title('Comparison of Model Performance (MAE and RMSE)')
ax.set_xticks(x)
ax.set_xticklabels(models)
ax.legend()

plt.show()
```
